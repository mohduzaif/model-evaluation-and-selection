### Import Libraries.

```python
import pandas as pd 
import numpy as np 
```

```python
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score, KFold
from sklearn.model_selection import GridSearchCV
```

### Overview of data.

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```python
df.sample(5)
```

```python
# separate the feature and target. 
X = df.iloc[ : , : 13]
y = df['medv']
```

```python
## Apply KNN Regressor without Grid Search CV.
```

```python
# Model object. 
knn_model1 = KNeighborsRegressor()

# kfold object. 
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# training using cross validation. 
scores = cross_val_score(knn_model1, X, y, cv=kfold, scoring='r2')

# scores. 
print('Mean of all the scores : ', np.mean(scores))
```

```python
## Apply using Grid Search CV. 
```

```python
# Model object. 
knn_model2 = KNeighborsRegressor()

# create the parameter dictionary. 
param_dict = {
    'n_neighbors' : [1, 3, 5, 7, 9, 8, 11, 10, 17, 16, 19],
    'weights' : ['uniform', 'distance'], 
    'algorithm' : ['ball_tree', 'kd_tree', 'brute'], 
    'p' : [1, 2]
}


# finding best parameters using grid search cv. 
gscv = GridSearchCV(knn_model2, param_grid=param_dict, cv=kfold, refit=True, verbose=2, scoring='r2')

# training.
gscv.fit(X, y)
```

```python
# best estimator. 
gscv.best_estimator_
```

```python
# best score. 
gscv.best_score_
```

```python
# best parameters.
gscv.best_params_
```

```python
# results of all the combination. 
all_results = pd.DataFrame(gscv.cv_results_)
all_results
```

```python
# find most important features from the all results.
all_results[['param_algorithm',	'param_n_neighbors',	'param_p', 'param_weights', 'mean_test_score']].sort_values('mean_test_score',ascending=False)
```
